```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('we_fn_use_c_marketing_customer_value_analysis.csv')
data.head()
```

```python
num_data = data.select_dtypes(include=[np.number])
num_data.head()
```

```python
for col in num_data:
    sns.histplot(x=col,data=num_data)

    plt.show()
```

```python
#None of them look like they have a normal dostribution, Number of policies and number of open complains have an exponential distribution.
```

```python
# log transform on balance and t_amount :
num_data_log = np.log(num_data)
```

```python
for col in num_data_log:
    sns.histplot(x=col,data=num_data_log)

    plt.show()
```

```python
#log-tranform changes the result, income, for example, is now skewed to the right instead of having a sort of plateau. The distribution of months since last claim has also changed greatly. Total claim amount has changed form being sked to the left to the right to being skewed to the left. 
```

```python

from sklearn.preprocessing import MinMaxScaler
# from sklearn.preprocessing import StandardScaler

# Scaling data
transformer = MinMaxScaler().fit(num_data) # need to keep transformer
num_data_normalized = transformer.transform(num_data)
num_data_norm = pd.DataFrame(num_data_normalized)
```

```python
for col in num_data_norm:
    sns.histplot(x=col,data=num_data_norm)

    plt.show()
```

```python
#this one simply changes the range (o -1) but not the distribution
```

```python
from sklearn.preprocessing import StandardScaler

# Scaling data
transformer1 = StandardScaler().fit(num_data) # need to keep transformer
num_data_normalized1 = transformer1.transform(num_data)
num_data_norm1 = pd.DataFrame(num_data_normalized1)
```

```python
for col in num_data_norm1:
    sns.histplot(x=col,data=num_data_norm1)

    plt.show()
```
